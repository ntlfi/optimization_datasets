```python
# read data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('potential_loc.csv')
```

```python
df
```

```python
data = pd.read_csv(f'../child_care_regulated/child_care_regulated.csv')
data['lat_lon'] = data[['latitude', 'longitude']].values.tolist()

data['zip_code'] = data['zip_code'].astype(str)

for i in range(len(data)):
    if len(data['zip_code'][i]) < 5:
        data.loc[i, 'zip_code'] = '0' * (5 - len(data['zip_code'][i])) + data['zip_code'][i]
    if len(data['zip_code'][i]) == 9:
        data.loc[i, 'zip_code'] = data['zip_code'][i][:5]

full_list_of_current_facilities = data['lat_lon'].tolist()
corresponding_zip_codes = data['zip_code'].astype(str).tolist()
```

```python
df
```

```python
def most_frequent(List):
    return max(set(List), key=List.count)

zip_codes = []
for i in range(len(df)):
    adjacent_facilities = []
    for j in range(len(full_list_of_current_facilities)): 
        if np.linalg.norm(np.array(df.iloc[i]) - np.array(full_list_of_current_facilities[j])) < 0.1: 
            adjacent_facilities.append(corresponding_zip_codes[j])

    zip_codes.append(most_frequent(adjacent_facilities))

    if i % 100 == 0:
        print(i)

df["zip_code"] = zip_codes
```

```python
zipcode_list = []
i = 0
zc = ''
while i < 50000:
    zc += zip_codes[i]

    if (i + 1) % 5 == 0:
        zipcode_list.append(zc)
        zc = ''

    i += 1
```

```python
df["zip_code"] = zipcode_list
```

```python
# write to csv
df.to_csv('potential_loc_with_zip.csv', index=False)
```
